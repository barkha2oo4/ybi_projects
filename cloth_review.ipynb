# Women Cloth Reviews Prediction with Multi Nomial Naive Bayes

the multinomial Naive Bayes classifier is suitable for classification with discrete features(e.g., word counts for text classification).The multinomial distribution normally requires integer feature counts. However, in practice,fractional counts such as tf-idf may also work

## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Dataset

```python
df= pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Teaching-Data/main/Women%20Clothing%20E-Commerce%20Review.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

## Missing Values

Remove missing values in Reviews colums with No Review test


```python
df.isna().sum()
```

```python
df[df['Review']==""]=np.NaN
```

```python
df['Review'].fillna("No Review",inplace=True)
```

```python
df.isna().sum()
```

```python
df['Review']
```

Define Target (y) and Feature (X)

```python
df.columns
```

```python
X=df['Review']
y=df['Rating']
```

```python
df['Rating'].value_counts()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.7,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## Get Feature Text Conversion to Tokens

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(lowercase=True,analyzer='word',ngram_range=(2,3),stop_words='english',max_features=5000)
```

```python
X_train=cv.fit_transform(X_train)
```

```python
cv.get_feature_names_out()
```

```python
X_train.toarray()
```

```python
X_test=cv.fit_transform(X_test)
```

```python
cv.get_feature_names_out()
```

```python
X_test.toarray()
```

## Get Model Train

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train, y_train)
```

## Get Model Prediction

```python
y_pred=model.predict(X_test)
y_pred.shape
```

```python
y_pred
```

## Get Probablity of Each Predicted Class

```python
model.predict_proba(X_test)
```

## Get Model Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

## Recategories Ratings as Poor (0) and Good(1)

```python
df['Rating'].value_counts()
```

```python
#RE-Rating as 1,2,3 as 0 and 4,5 as 1
df.replace({'Rating': {1:0,2:0,3:0,4:1,5:1}}, inplace = True)
```

```python
y=df['Rating']
X=df['Review']
```

### Train Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,stratify=y,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

### Get Feature Text Conversion to Tokens

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(lowercase=True,analyzer='word',ngram_range=(2,3),stop_words='english',max_features=5000)
```

```python
X_train=cv.fit_transform(X_train)
X_test=cv.transform(X_test)
```

### Get Model Re-Train

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train, y_train)
```

### Get Model Prediction

```python
y_pred=model.predict(X_test)
y_pred.shape
```

```python
y_pred
```

### Get Model Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
